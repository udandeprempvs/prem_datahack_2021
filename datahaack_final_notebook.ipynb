```python
import pandas as pd
df = pd.read_csv("/content/numadic-customer-transaction-prediction/train.csv")
df.head()
```

```python
df.describe()
```

```python
df_0=df.drop(df[df['target']==1].index,inplace=False,axis=0)
```

```python
df_0.head()
```

```python
df_frac=df_0.sample(n = 20000,replace = False,random_state=0)
(df_frac.target==1).value_counts()
```

```python
df_1=df.drop(df[df['target']==0].index,inplace=False,axis=0)
```

```python
df_0.describe()
```

```python
main = pd.concat([df_frac,df_1],axis=0)
main.describe()
```

```python
x = main.drop(['target'],axis=1)
y = main.target
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(units = 6, activation = 'relu', input_dim = 200))

model.add(Dense(units = 6, activation = 'relu'))

model.add(Dense(units = 1, activation = 'sigmoid'))

model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

model.summary()
```

```python
model.fit(x_train, y_train, batch_size = 64, epochs = 30)
```

```python
score, acc = model.evaluate(x_train, y_train,
                            batch_size=10)
print('Train score:', score)
print('Train accuracy:', acc)

print('*'*20)

# Evaluating on Test data
score, acc = model.evaluate(x_test, y_test,
                            batch_size=10)
print('Test score:', score)
print('Test accuracy:', acc)
```
